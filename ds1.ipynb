```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
np.random.seed(42)

population_marks = np.random.normal(loc=60, scale=10, size=1000)
```

```python
sample_marks = np.random.choice(population_marks, size=200, replace=False)
```

```python
df = pd.DataFrame(sample_marks, columns=["Marks"])
```

```python
mean = df["Marks"].mean()
std_dev = df["Marks"].std()

print("Sample Mean:", mean)
print("Sample Standard Deviation:", std_dev)
```

```python
sns.histplot(df["Marks"], kde=True)
plt.title("Gaussian Distribution of Sample Marks")
plt.show()
```

```python
stat, p_value = stats.shapiro(df["Marks"])

print("P-value:", p_value)
```

```python
df["Z_score"] = (df["Marks"] - mean) / std_dev
```

```python
outliers = df[np.abs(df["Z_score"]) > 3]
print(outliers)
```
